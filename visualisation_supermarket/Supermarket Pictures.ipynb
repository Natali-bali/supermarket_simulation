```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
import numpy as np
```

```python
im = Image.open('supermarket.png')
market = np.array(im)
print(market.shape, market.dtype)
```

#### I will make another image with 4 locations and names for my visualisation

```python
market = np.delete(market, np.arange(544,576), axis=1)
```

```python
market.shape
```

```python
blue_lines = market[0 : 32*13, 0:32*2]
```

```python
Image.fromarray(blue_lines )
```

```python
sriped_line = market[0 : 32*13, 5*32:5*32+32*7]
```

```python
Image.fromarray(sriped_line)
```

```python
market = np.append(market,sriped_line, axis=1)
```

```python
market = np.append(market,blue_lines, axis=1)
```

```python
market.shape
```

```python
Image.fromarray(market)
```

```python
blue_str = market[0 : 8*32, 0:2*32]
```

```python
blue_str.shape
```

```python
Image.fromarray(blue_str)
```

```python
market[2*32:10*32, 4*32:6*32] = blue_str
market[2*32:10*32, 8*32:10*32] = blue_str
market[2*32:10*32, 12*32:14*32] = blue_str
market[2*32:10*32, 16*32:18*32] = blue_str
market[2*32:10*32, 20*32:22*32] = blue_str
```

```python
one_tile = market[3*32:4*32, 2*32:3*32]
```

```python
market[1*32:2*32, 16*32:17*32] = one_tile
```

```python
img = Image.fromarray(market)
```

```python
draw = ImageDraw.Draw(img)
font = ImageFont.truetype("Banny-Round.ttf", 24)
draw.text((5*32, 0),"SUPERMARKET VISUALISATION",(255,255,255),font=font)
```

'checkout','dairy', 'drinks', 'fruit', 'spices'

```python
font = ImageFont.truetype("Banny-Round.ttf", 20)
draw.text((4*32, 1*32), "CHECKOUT",  font=font, fill=255)
draw.text((8*32, 1*32+4), "DAIRY",  font=font, fill=255)
draw.text((12*32, 1*32+4), "DRINKS",  font=font, fill=255)
draw.text((16*32, 1*32+4), "FRUIT",  font=font, fill=255)
draw.text((20*32, 1*32+4), "SPICES",  font=font, fill=255)
```
