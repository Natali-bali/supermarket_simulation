## Simulation of customers behavior

To create a proper simulation we have to follow the pattern of previous simulation. So we need to find:
- probability of entering first location
- how many clients enter the shop every minute (we should be able to increase/decrease this number, to simulate different situation)
- probability of entering next location
- how much time clients spend at each location

```python
df = pd.read_csv('xlsx/clean_xl_final.csv', delimiter = ';', index_col = 'timestamp')
df = df.iloc[1:]
```

```python
df_1 = df.fillna('checkout')
df_1 = df_1.loc[df_1['step'] != 1]
```

```python
matrix = pd.crosstab(df_1['location'], df_1['next'], normalize=0)
```

### Create the vectors of probabilities to spend certain time at each location

```python
sect={}
for loc in df['location'].unique():
    a = []
    deltas = np.array(df.loc[df['location'] == loc, 'delta'].unique())
    deltas = list(np.sort(deltas))
    for delta in deltas:
        e = df.loc[(df['location'] == loc) & (df['delta']==delta), 'delta'].count()/df.loc[df['location']==loc, 
                                                                                           'delta'].count()
        a.append(e)
    sect[loc] = [deltas, a]
```

```python
class Customer:
    
    client_pic = cv2.imread('pics/customer.png')
    
    def __init__(self, customer_id, location, matrix = matrix, ll = locations, sect = sect, x = 0, y = 0):
        self.customer_id = customer_id #provided by Supermarket class
        self.location = location #number enter location provided by Supermarket class
        self.matrix = matrix
        self.ll = ll
        self.sect = sect
        self.time_count = self.time_next()
        self.x = x
        self.y = y
        
        
    def __repr__(self):
        return f"The customer {self.customer_id} is at {self.ll[self.location]} and will spend there {self.time_count} minutes" 
        
    def is_not_active(self):
        return self.location == 0

    def time_next (self):  
        loc_name = self.ll[self.location]
        t = int(np.random.choice(self.sect[loc_name][0], 1, p = self.sect[loc_name][1]))
        return t

    def state(self):
        if self.time_count != 1:
            self.time_count = self.time_count - 1
        else:
            self.next_state()
    
    def next_state(self):
        self.time_count = self.time_next()
        location = self.location
        matrix = self.matrix
        ll = self.ll
        p = list(matrix.loc[ll[location]])
        self.location = ll.index(np.random.choice(ll, 1, p = p))
        self.x = 160 + (self.location)*128
     
    #     Drawing function
    def draw(self, frame):
        """draw into frames"""
        frame[self.y:self.y+32, self.x:self.x+32] = self.client_pic
        return frame   
        
```

Class Supermarket manages multiple Customer instances that are currently in the market.

```python
class Supermarket:    
    def __init__(self, n_time, p_enter=p_enter, cl_ent = cl_ent, ll=locations):        
        self.customers = []
        self.last_id = 0
        self.n_time = n_time
        self.last_id = 1
        self.time_count = 0 
        self.p_enter = p_enter
        self.cl_ent = cl_ent 
        self.time_now = 0
        self.hour = 0
        self.minutes = 0
        self.result = {} #dict of resuts for output
        self.count = 0 #count steps in print customers
        self.ll=ll

#     def __repr__(self):
#         pass

    def get_time(self):
        hour = 7 + self.minutes // 60
        if hour == 22:
            hour = 7 
            self.minutes = 0
            # If new day starts, we have to remove all customers from shop and start again
            self.remove_all_customers()
        mi = self.minutes % 60
        self.hour = hour - 7 #for our array cl_ent indexed from 0
        self.time_now = f"{hour:02}:{mi:02}:00"

    def print_customers(self):
#         print(self.time_now) #To see printed output uncomment these 
#         print(self.customers) #To see printed output uncomment these 
        for c in self.customers:
            self.count = self.count + 1
            self.result[self.count] = {'time': self.time_now,
                                        'customer_id':c.customer_id,
                                        'location':self.ll[c.location],
                                        'delta': c.time_count}       
        
    def next_minute(self):
        self.add_new_customers()
        self.get_time()
        print(self.print_customers())        
        for t in range(1, self.n_time + 1): #when we activate class we give argument minutes
            self.minutes += 1
            self.time_count += 1
            self.get_time()
            self.move()
            self.add_new_customers()            
            print(self.print_customers())
            self.remove_exit_customers()
                
    def add_new_customers(self):  #Create customers
        for i, h in enumerate(cl_ent):
            if i == self.hour:
                cust_count = int(np.random.poisson(h*1.3)) 
        for n in range(cust_count):
            #we use probability vector to choose first location
            location = int(np.random.choice([1, 2, 3, 4], 1, p = p_enter))
            self.customers.append(Customer(self.last_id, location, x=128+(location)*128, y=64+n*32))
            self.last_id += 1
            self.show()
            
    def show(self):
        frame = background.copy()
        for cust in self.customers:
            frame = cust.draw(frame)
#         cv2.imshow('frame', frame) uncomment to see real time visualisation
#  I save png to make gif file. For this purpose i would run 240 min, otherwise gif will be too big.
#         cv2.imwrite(f'output_png/{self.time_count}.png', frame)
        cv2.waitKey(1)
        cv2.destroyAllWindows()        
    
    def move(self):    #Move customers
        if self.customers != []:
            for c in self.customers:
                c.state()         

    def remove_exit_customers(self): #removes every customer that is not active any more.
        for cust in self.customers:
            if cust.is_not_active():
#                 print("This customer will be removed: ", cust.customer_id)
                self.customers.remove(cust)
#         I cant explain this bug, but not all customers removed from check out after first loop. So i repeat
#         this loop again, now all customers removed from check out
        for cust in self.customers:
            if cust.is_not_active():
#                 print("Second---This customer will be removed: ", cust.customer_id)
                self.customers.remove(cust)         
    
    def remove_all_customers(self):           
#         print("All customers will be removed without reaching checkout")
        self.customers= []
```

Simulate 4 days, 15 hours * 60 min  = 3600 minutes
